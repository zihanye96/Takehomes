# Overview

### About the Data

Each xml file contains credit information for a given customer. For example, a customer can have 3 different loans on file, and for each loan there's information such as whether or not the loan was paid on time for a given month (on time, 30 days late, 60+ days late), as well as other information like the loan type, the status of the loan (paid, current, late, etc), and the balance remaining on the loan. The duration of the loans vary, so one loan may only take 12 months to pay off but another can take 22 months.

### Structuring the Data

Here's how I plan to format the dataset:

Each row will be a customer. It will contain their SSN and ID (to uniquely identify the customer), credit score, as well as aggregate features for each customer based on their loan data. To automate the feature engineering process, I will use the featuretools package, which automates deep feature synthesis for relational data. For more details, please review the package [documentation](https://docs.featuretools.com/en/stable/automated_feature_engineering/afe.html).

### Model Fitting and Interpretation
After creating the dataset, I will use 2 models to measure performance. I will fit a random forest (tuned via 5-fold cv) and a gradient boosting machine (GBM) and use the better performing model as my final model. Tree-based models like random forests and gradient boosted trees can give me a sense of feature importance (via mean decrease impurity). So, from either model, I will be able to extrapolate the most important features (as deemed by the model) for determining one's credit score.

```python
## import required modules 
import numpy as np
import pandas as pd
import os
import xml.etree.ElementTree as ET
import warnings
warnings.filterwarnings('ignore')

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

import featuretools as ft
from sklearn import preprocessing

import statsmodels.api as sm
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
## get a list of all the filesnames in validation set
filenames_training = [name for name in os.listdir("./training")]

def parse_report_training(filename):

    tree = ET.parse("./training/" + filename)
    root = tree.getroot()

    ## extract id, ssn, and credit score
    id_num = root[0].text
    ssn = root[2].text
    credit_score = root[4].text

    ## initialize dataframe - each row in this df will represent one loan 
    tmp = pd.DataFrame()

    ## loop over the loans and save the extracted information
    for node in tree.iter('tradeline'): # each node corresponds to a loan

        i = 0
        info = {}
        for elem in node.iter(): # each element corresponds to a piece of info about a given loan

            if not elem.tag==node.tag and elem.tag not in ["payment", "payments", "month"]:
                if elem.tag == "status":
                    info.update({"{}_{}".format(elem.tag, i) : elem.text})
                    i += 1
                else:
                    info.update( {"{}".format(elem.tag) : elem.text})
            info.update({"n_months": i-1})
        df = pd.DataFrame(info, index=[0])

        ## if a loan is currently late, there's an extra tag called "overdue"
        if 'overdue' not in df.columns:
            df["overdue"] = 0
            cols = df.columns.tolist()
            cols = cols[-1:] + cols[:-1]
            df = df[cols]


        ## starting from column index 8, each column is whether or not payment was made on time for a given month
        ## the code below joins the first 8 columns for a loan with the proportions of on time and late payments 
        summary = df.iloc[:,:8].join(pd.DataFrame(df.iloc[0,8:].value_counts(normalize=True)).T)

        ## if no payments were made on time/30 days late/60+ days late, then impute the respective column with 0
        if '30 DAYS LATE' not in summary.columns:
            summary["30 DAYS LATE"] = 0
        if '60 DAYS LATE OR MORE' not in summary.columns:
            summary["60 DAYS LATE OR MORE"] = 0
        if 'ON TIME' not in summary.columns:
            summary["ON TIME"] = 0      



        tmp = tmp.append(summary, ignore_index=True)

    tmp = tmp.fillna(0)

    tmp["id"] = id_num
    tmp["ssn"] = ssn
    tmp["credit_score"] = credit_score

    ## convert columns to from text to numeric (if applicable)
    tmp["credit_score"] = tmp["credit_score"].astype(int)
    tmp["balance"] = tmp["balance"].astype(int)
    tmp["high_credit"] = tmp["high_credit"].astype(int)
    tmp["last_activity"] = tmp["last_activity"].astype(int)
    tmp["overdue"] = tmp["overdue"].astype(int)

    return tmp
```

```python
## parse every xml in training set and save the aggregated data as a df
model_data = pd.DataFrame()
for filename in filenames_training:
    summary_df = parse_report_training(filename)
    model_data = model_data.append(summary_df)

model_data.rename(columns={"30 DAYS LATE": "prop_30_late", "60 DAYS LATE OR MORE": "prop_60_late",
                       "ON TIME": "prop_on_time", "status_0": "status"}, inplace=True)
```

```python
## encode "active" to a 0-1 variable (0 if false)
model_data["active"] = np.where(model_data["active"] == "False", 0, 1)

## create dummy variables for categorical features (i.e. status, type)
def create_dummy(df, feature_names):
    dummydf = pd.DataFrame()
    for name in feature_names:
        tmp = pd.get_dummies(df.loc[:,name], prefix= name, prefix_sep='_')
        dummydf = pd.concat([dummydf, tmp], axis = 1)
    return dummydf

categ_features = ["status", "type"]
quant_features = [x for x in model_data.columns if x not in categ_features]

quant_df = model_data.loc[:, quant_features]
dummy_df = create_dummy(model_data, categ_features)

model_data = pd.concat([quant_df, dummy_df], axis = 1)
```

```python
model_data.head()
```

```python
## split original dataset into 3 (id, credit score, and everything else)
ids = model_data["id"].unique()
target = model_data.groupby('id').agg({'credit_score':'mean'})
data = model_data.drop(['ssn', 'credit_score'], axis=1)

## create a df of just the unique ids
customer_ids = pd.DataFrame({"id": ids})
```

```python
## automated feature engineering using featuretools
es = ft.EntitySet(id = 'customer')
es = es.entity_from_dataframe(entity_id = 'customer_id', dataframe = customer_ids, index = 'id')
es = es.entity_from_dataframe(entity_id = 'loan_info', dataframe = data,
                              make_index = True, index = 'loan_index')

## link the id dataframe with the model_data dataframe (so we can aggregate over id)
r_id_loan = ft.Relationship(es['customer_id']['id'], es['loan_info']['id'])

es = es.add_relationships([r_id_loan])
es
```

```python
## define the set of primitive features used for aggregation and generate aggregate features
agg_primitives =  ["sum", "std", "max", "skew", "min", "mean", "count", "mode"]
feature_matrix_spec, feature_names_spec = ft.dfs(entityset = es, target_entity = 'customer_id', 
                                                 agg_primitives=agg_primitives, max_depth = 3, features_only=False,
                                                 verbose=True)
```

```python
## merge the credit scores and the aggregated dataset
data = pd.merge(target, feature_matrix_spec, how='inner', left_on = 'id', right_on = 'id').reset_index()
data.head()
```

```python
Y = data.loc[:,"credit_score"]
X = data.drop(['id', 'credit_score'], axis=1)
```

```python
print("The mean credit score in the training set is:", np.mean(Y))
print("With a standard deviation of:", np.std(Y))
```

```python
sns.distplot(Y, bins = 16)
plt.title("Distribution of Credit Scores")
```

The credit scores looks normally distributed and centered around 1000, with a standard deviation of about 350. This means we don't have to transform the credit scores to be normally distributed (ex. box-cox) and we now have a sense of how good/bad our prediction errors will be.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)

## a function to compute the rmse
def compute_rmse(predictions, yvalues):
    return np.sqrt(sum((np.array(predictions)-np.array(yvalues))**2)/len(yvalues))
```

```python
## initialize the model and parameter space

rf_model = RandomForestRegressor(random_state = 5, max_features = 'sqrt')
parameters = {'n_estimators': np.arange(30, 301, 30)}

np.random.seed(seed=1000)
rf_gridcv = GridSearchCV(estimator = rf_model, param_grid = parameters, cv = 5, scoring = 'neg_mean_squared_error')
rf_gridcv.fit(X_train, Y_train)
print("The optimal choice of hyperparameters is:", rf_gridcv.best_params_)
print("With a negative mse of:", rf_gridcv.best_score_)
```

```python
## fit a random forest based on the optimal hyperparameters 

mod = RandomForestRegressor(n_estimators = 300)
mod.fit(X_train, Y_train)
rf_preds = mod.predict(X_test)
compute_rmse(rf_preds, Y_test)
```

```python
## note: I didn't tune the GBM since it was too computationally intensive

gb_model = GradientBoostingRegressor(learning_rate = .1, n_estimators = 500)
gb_model.fit(X_train, Y_train)
gbm_preds = gb_model.predict(X_test)
compute_rmse(gbm_preds, Y_test)
```

```python
## get a sense of feature importance based on random forest
importances = mod.feature_importances_
most_important_idx = importances.argsort()[-10:]
features = X_train.columns
```

```python
plt.title('Feature Importances (Random Forest)')
plt.barh(range(len(most_important_idx)), importances[most_important_idx], color='lightsteelblue', align='center')
plt.yticks(range(len(most_important_idx)), [features[i] for i in most_important_idx])
plt.xlabel('Relative Importance')
plt.show()
```

### Results

The random forest model yielded a test RMSE of about 65 while the GBM yielded a test RMSE of about 56. Given the scale of the distribution of credit scores ($\mu \approx 1000$, $\sigma \approx 350$), the initial models performed moderately well, though there is certainly much room for improvement. 

From the GBM, we can see that the most important features for determining credit score are the total amount of money borrowed across all loans. Other important features include the standard deviation of number of late loans, the proportion of late loans, and whether or not the loan status was "collections". So, if each person has $1,000 to improve their credit score, they should use it to protect themselves against late payments or defaults (after which the loan goes into "collection" status).
